```python
%matplotlib inline
import matplotlib.pyplot as plt

```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0,45)
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
plt.bar(x_axis, y_axis, color='magenta', label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_xlabel('Fare ($)')
plt.gca().invert_yaxis()
ax.legend()
```

```python
plt.plot(y_axis, x_axis, 'o', color='red')
```

```python
fig, ax = plt.subplots()
ax.plot(y_axis, x_axis, 'o', color='red', label='Chicago')
ax.set_xlabel('Fare($)')
ax.set_ylabel('')
plt.gca().invert_yaxis()
ax.legend()
plt.title('PyBer Fare by Month')
plt.xlim(min(y_axis), max(y_axis) + 5)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,s = [i*5 for i in y_axis], alpha=.2, linewidths=2, label='Boston', edgecolors='Black')
plt.gca().invert_yaxis()
ax.legend()
plt.title('PyBer Fare by Month')
ax.set_xlim(min(y_axis) - 5, max(y_axis) + 5)
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', colors=colors);
```

```python
fig, ax = plt.subplots()
explode_values = (0, 0, 0.5, 0, 0, 0, 0.5, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', colors=colors, explode=explode_values, shadow=True, counterclock=True, startangle=60)
plt.show()
```
